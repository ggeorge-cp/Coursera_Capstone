```python
# Import pandas to create pandas dataframe
import pandas as pd
```

```python
# Read in the wikipedia table as a pandas dataframe 
wiki_data = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')[0]

wiki_data.head()
```

```python
# Remove any rows where the Borough is "Not assigned"
wiki_data = wiki_data[wiki_data["Borough"] != "Not assigned"]

wiki_data.reset_index(drop=True, inplace=True)

wiki_data.head()
```

```python
# Display the shape of the dataframe
wiki_data.shape
```

```python
#import geocoder
import geocoder
```

```python
for i in range(0, len(wiki_data.index)):
    lat_lng_coords = None
    
    postal_code = wiki_data.iloc[i]['Postal Code']
    
    while(lat_lng_coords is None):
        g = geocoder.google('{}, Toronto, Ontario'.format(postal_code))
        lat_lng_coords = g.latlng
    
    latitude = lat_lng_coords[0]
    longitude = lat_lng_coords[1]
    
    wiki_data.iloc[i]['Latitude'] = latitude
    wiki_data.iloc[i]['Longitude'] = longitude
    
wiki_data
```
